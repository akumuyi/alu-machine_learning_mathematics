```python
# YOUR CODE HERE
"""
Experimenting with bin size 20
"""
import numpy as np

co2_data = df["co2"]
plt.figure(figsize=(10, 6))
plt.hist(co2_data, bins=30, edgecolor='black')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Number of Vehicles')
plt.title('Distribution of CO2 Emissions')
plt.show()
```

```python
# Experimenting with different bin sizes

plt.figure(figsize=(10, 6))
plt.hist(co2_data, bins=15, edgecolor='black')
plt.title('Distribution of CO2 Emissions with 15 Bins')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(co2_data, bins=50, edgecolor='black')
plt.title('Distribution of CO2 Emissions with 50 Bins')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.show()
```

```python
# YOUR CODE HERE
comb_data = df["comb"]
plt.figure(figsize=(10, 6))
plt.hist(comb_data, bins=30, edgecolor='black')
plt.title('Distribution of Combined Fuel Consumption')
plt.xlabel('Combined Fuel Consumption (mpg or L/100km)')
plt.ylabel('Frequency')
plt.show()
```

```python
# YOUR CODE HERE
displ_data = df["displ"]
plt.figure(figsize=(10, 6))
plt.hist(displ_data, bins=30, edgecolor='black')
plt.title('Distribution of Engine Displacement')
plt.xlabel('Engine Displacement (Liters)')
plt.ylabel('Frequency')
plt.show()
```

```python
numerical_cols = df.select_dtypes(include='number')
correlation_matrix = numerical_cols.corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, fmt='.2f', cmap='coolwarm', cbar=True)
plt.show()
```

```python
target_correlation = correlation_matrix['ghgScore'].abs()
columns_to_drop = target_correlation[target_correlation < 0.1].index.tolist()

print("Columns to consider dropping:", columns_to_drop)
```

**Justification**:

*Low Correlation:* Columns with very low correlation (close to 0) contribute little to predicting ghgScore and can be dropped to reduce model complexity.

*Multicollinearity:* If two columns are highly correlated with each other but not with ghgScore, one of them can be dropped.

### Interpretation of the Histograms:

#### 1. **CO2 Emissions Distribution**
   - The distribution of CO2 emissions (first histogram) appears to be **right-skewed** (positively skewed), with a concentration of vehicles emitting lower levels of CO2.
   - Most vehicles fall within the lower CO2 emission ranges, particularly between 200 to 400 g/km.
   - There are fewer vehicles with very high CO2 emissions, but a long tail extends toward the right, indicating some vehicles with significantly higher emissions.

   - **Bin Size Impact**:
     - A smaller bin size (50 bins) provides a more detailed view of the distribution, showing finer variations in frequency.
     - A larger bin size (15 bins) smooths the data, making it easier to see the general trend but potentially missing some of the finer details.

#### 2. **Combined Fuel Consumption Distribution**
   - The combined fuel consumption histogram also shows a **right-skewed** distribution.
   - Most vehicles have a combined fuel consumption in the lower range, indicating that a significant number of vehicles are fuel-efficient.
   - The frequency decreases as fuel consumption increases, with fewer vehicles consuming higher amounts of fuel.

   - The right skew suggests that while fuel efficiency is a priority for most vehicles, there are still some vehicles that consume a lot more fuel, possibly due to larger engines or less efficient technologies.

#### 3. **Engine Displacement Distribution**
   - The engine displacement histogram shows a **bimodal distribution**, with two noticeable peaks.
   - The first peak is around the 2.0-liter range, indicating a large number of vehicles have engines with around 2 liters of displacement. This is common for many compact and midsize cars.
   - The second, smaller peak is around the 3.5-4.0 liter range, which could represent larger vehicles, like SUVs and trucks, or high-performance cars.
   - The drop-off after 4.0 liters suggests that very large engines are less common in this dataset.

   - This bimodal distribution reflects the diversity in vehicle types within the dataset, ranging from smaller, more fuel-efficient engines to larger, performance-oriented or utility-focused engines.
